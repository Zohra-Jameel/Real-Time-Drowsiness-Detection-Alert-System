```python
import uuid   # Unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['awake', 'drowsy']
number_imgs = 5
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2
import time
import threading
from playsound import playsound

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='best.pt', force_reload=True)

# Start webcam
cap = cv2.VideoCapture(0)

drowsy_start = None
alarm_playing = False


def play_alarm():
    global alarm_playing
    alarm_playing = True
    playsound("alarm.mpeg")
    alarm_playing = False


while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Run YOLO detection
    results = model(frame)

    labels = results.xyxyn[0][:, -1]
    coords = results.xyxyn[0][:, :-1]

    for i in range(len(labels)):
        label = model.names[int(labels[i])]
        row = coords[i]

        if row[4] >= 0.5:

            x1 = int(row[0] * frame.shape[1])
            y1 = int(row[1] * frame.shape[0])
            x2 = int(row[2] * frame.shape[1])
            y2 = int(row[3] * frame.shape[0])

            # Draw bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
            cv2.putText(frame, label, (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

            if label == "drowsy":

                if drowsy_start is None:
                    drowsy_start = time.time()

                elif time.time() - drowsy_start >= 2:
                    if not alarm_playing:
                        threading.Thread(target=play_alarm).start()

            else:
                drowsy_start = None

    cv2.imshow("Drowsiness Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
